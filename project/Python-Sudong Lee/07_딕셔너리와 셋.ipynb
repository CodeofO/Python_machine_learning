## ë”•ì…”ë„ˆë¦¬ ì»´í”„ë¦¬í—¨ì…˜(Dictionalry Comprehension)
ë¦¬ìŠ¤íŠ¸ì™€ ê°™ì´ ë”•ì…”ë„ˆë¦¬ë„ ì»´í”„ë¦¬í—¨ì…˜ ê¸°ëŠ¥ì„ ê°€ì§€ê³  ìˆìŠµë‹ˆë‹¤. <br>
`{key_expression : value_expression for expression in iterable}`

```python
word = 'letters'
for letter in word:
  print(word.count(letter)) # ê°ê°ì˜ ì›ì†Œì˜ ê°œìˆ˜ë¥¼ word ì•ˆì—ì„œ countí•œ ê°’ì„ ë‚´ë³´ëƒ„
```

```python
word = 'letters'
letter_counts = {letter: word.count(letter) for letter in word}
letter_counts
```

```python
vowels = 'aeiou'
word = 'onomatopoeia'
vowel_counts = {letter: word.count(letter) for letter in word if letter in vowels} # ê°€ì¥ íŒŒì´ì¬ ë‹¤ìš´ ì½”ë”©
vowel_counts
```
