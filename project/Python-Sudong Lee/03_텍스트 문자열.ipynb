## ìŠ¬ë¼ì´ì‹±(Slicing)
- ìŠ¬ë¼ì´ì‹±ì€ ì‚¬ìš©í•˜ì—¬ í•œ ì‹œí€€ìŠ¤(ë¬¸ìì—´)ì—ì„œ ì—¬ëŸ¬ ìš”ì†Œ(ë¬¸ì)ë¡œ ì´ë£¨ì–´ì§„ **ìŠ¬ë¼ì´ìŠ¤(slice)**ë¥¼ ì¶”ì¶œí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
- ëŒ€ê´„í˜¸(`[]`)ë¥¼ ì‚¬ìš©í•˜ì—¬ ì‹œì‘(`start`) ì˜¤í”„ì…‹, ë(`end`) ì˜¤í”„ì…‹, ì˜µì…˜ìœ¼ë¡œ ìŠ¤í…(`step`)ì„ ëª…ì‹œí•˜ì—¬ ìŠ¬ë¼ì´ìŠ¤ë¥¼ ì •ì˜í•©ë‹ˆë‹¤. (ì¼ë¶€ë¥¼ ìƒëµí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.)
- ìŠ¬ë¼ì´ìŠ¤ëŠ” ì˜¤í”„ì…‹ì˜ `start`ì™€ `end-1` ì‚¬ì´ ë¬¸ìë¥¼ í¬í•¨í•©ë‹ˆë‹¤.

ì•„ë˜ì˜ ìŠ¬ë¼ì´ì‹± ê·œì¹™ì„ ì‚´í´ë´…ì‹œë‹¤. 
- `[:]`: ì²˜ìŒë¶€í„° ëê¹Œì§€ ì „ì²´ ì‹œí€€ìŠ¤ë¥¼ ì¶”ì¶œí•œë‹¤.
- `[start:]`: `start` ì˜¤í”„ì…‹ë¶€í„° ëê¹Œì§€ ì‹œí€€ìŠ¤ë¥¼ ì¶”ì¶œí•œë‹¤.
- `[:end]`: ì²˜ìŒë¶€í„° `(end - 1)` ì˜¤í”„ì…‹ê¹Œì§€ ì‹œí€€ìŠ¤ë¥¼ ì¶”ì¶œí•œë‹¤.
- `[start : end]`: `start` ì˜¤í”„ì…‹ë¶€í„° `(end - 1)` ì˜¤í”„ì…‹ê¹Œì§€ ì‹œí€€ìŠ¤ë¥¼ ì¶”ì¶œí•œë‹¤.
- `[start : end : step]`: `step`ë§Œí¼ ë¬¸ìë¥¼ ê±´ë„ˆë›°ë©´ì„œ, `start` ì˜¤í”„ì…‹ë¶€í„° `(end - 1)` ì˜¤í”„ì…‹ê¹Œì§€ ì‹œí€€ìŠ¤ë¥¼ ì¶”ì¶œí•œë‹¤.

````python
print('letters[:]\t', letters[:])
````

````python
print('letters[20:]\t', letters[20:])
print('letters[10:]\t', letters[10:])
print('letters[:20]\t', letters[:20])
print('letters[:10]\t', letters[:10])
````
