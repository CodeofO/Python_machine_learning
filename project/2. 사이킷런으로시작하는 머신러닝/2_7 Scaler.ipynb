```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
```

```python
iris = load_iris()
data = iris.data
iris_df = pd.DataFrame(data=data, columns=iris.feature_names)


scaler = StandardScaler()
scaler.fit(iris_df)

iris_scaled = scaler.transform(data)
iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)
mean_df = pd.DataFrame(iris_df_scaled.mean())
```

```python
Mscaler = MinMaxScaler()
Mscaler.fit(data)

# df => 
iris_Mscaled = Mscaler.transform(iris_df)

iris_df_Mscaled = pd.DataFrame(data=iris_Mscaled, columns=iris.feature_names)

iris_df_Mscaled.max()
```

```python
train_array = np.arange(0, 11).reshape(-1, 1)
test_array = np.arange(0, 6).reshape(-1, 1)

Mscaler.fit(train_array)

train_scaled = Mscaler.transform(train_array)

print('origin_train :',train_array.reshape(-1))
print('scaled_train :',np.round(train_scaled.reshape(-1),2))

# scaled_test(1)
Mscaler2 = Mscaler.fit(test_array)
test_scaled1 = Mscaler2.transform(test_array)

# scaled_test(2)
Mscaler.fit(train_array)
test_scaled2 = Mscaler.transform(test_array)

print('test_scaled1 :', test_scaled1.reshape(-1))
print('test_scaled2 :', test_scaled2.reshape(-1))
```
