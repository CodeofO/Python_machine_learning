# 2.붓꽃 품종 예측하기

1️⃣ 데이터 세트 분리: 데이터를 학습 데이터와 테스트 데이터로 분리합니다. 

```python
iris = load_iris()

# iris.data는 Iris data set에서 feature만으로 된 데이터를 numpy로 가지고 있다
iris_data = iris.data
iris_label = iris.target

# iris.target은 붓꽃 데이터 세트에서 레이블(결정 값)데이터를 numpy로 가지고 있다
print('iris target값:', iris_label)
print('iris taget명:', iris.target_names)

#붗꽃 data set를 자세히 보기 위해 DataFrame으로 변환
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['label']=iris.target
iris_df.head(3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)

# First parameter : iris_data = feature data set
# Second parameter : iris_label = Lable data set
# Third parameter : test_size = scope of data set
# Fourth parameter : 호출 시 같은 학습/데이터 용 데이터 세트를 생성하기 위한 난수발생기 
```

2️⃣ 모델 학습: 학습 데이터를 기반으로 ML 알고리즘을 적용해 모델을 학습시킵니다.

```python
# DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state=11)

# 학습 수행
dt_clf.fit(X_train, y_train)

iris_df
```

3️⃣ 예측 수행: 학습된 ML 모델을 이용해 테스트 데이터의 분류(즉,붓꽃종류)를 예측합니다.

```python
# 학습이 완료된 DecisionTreeClassifier 객체에서 테스트 데이터 세트로 예측 수행
pred = dt_clf.predict(X_test)
```

4️⃣ 평가 이렇게 예측된 결팟값과 테스트 데이터의 실제 결갓값을 비교해 ML 모델 성능을 평가합니다.

```python
from sklearn.metrics import accuracy_score
print('예측 정확도:{0: .4f}'.format(accuracy_score(y_test, pred)))
```

## 교차 검증

### K 폴드 교차 검증

https://www.notion.so/0783d3c758b946cbba493023c47133f6

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features = iris.data
label = iris.target
dt_clf = DecisionTreeClassifier(random_state=156)

# 5개의 폴드 세트로 분리하는 KFold 객체와 폴드 세트별 정확도를 담을 리스트 객체 생성.
kfold = KFold(n_splits=5)
cv_accuracy = []
print('붗꽃 데이터 세트 크기:', features.shape[0])
```

```python
n_iter = 0

# KFold 객체 split()를 호출하면 폴드 별 학습용, 검증용 테스트의 로우 인덱스를 array로 반환
for train_index, test_index in kfold.split(features):
    # kfold.split() 으로 반환된 인덱스를 이용해 학습용, 검증용 테스트 데이터 추출
    x_train, x_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    
    # 학습 및 예측
    dt_clf.fit(x_train, y_train)
    pred = dt_clf.predict(x_test)
    n_iter += 1
    
    # 반복 시마다 정확도 측정
    accuracy = np.round(accuracy_score(y_test, pred), 4)
    train_size = x_train.shape[0]
    test_size = x_test.shape[0]
    print('\n#{0} 교차검증 정확도: {1}, 학습데이터 크기: {2}, 검증데이터 크게: {3}'.format(n_iter, accuracy, train_size, test_size))
    print('#{0} 검증 세트 인덱스: {1}'.format(n_iter, test_index))
    cv_accuracy.append(accuracy)

    
print('\n## 평균 검증 정확도:', np.mean(cv_accuracy))
```

### Stratified K 폴드

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df['label'].value_counts()
```

```python
kfold = KFold(n_splits = 3)
n_iter = 0
for train_index, test_index in kfold.split(iris_df):
    n_iter += 1
    label_train = iris_df['label'].iloc[train_index]
    label_test = iris_df['label'].iloc[test_index]
    print('\n\n## 교차 검증: {0} ##'.format(n_iter))
    
    print('label_train data distribution:\n', label_train.value_counts())
    print('\nlabel_test data distribution:\n', label_test.value_counts())
```

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=3)
n_iter=0

for train_index, test_index in skf.split(iris_df, iris_df['label']):
    n_iter += 1
    label_train = iris_df['label'].iloc[train_index]
    label_test = iris_df['label'].iloc[test_index]
    print('## 교차 검증: {0}'.format(n_iter))
    print('label_train data distribution:\n', label_train.value_counts())
    print('label_test data distribution:\n', label_test.value_counts(),'\n\n')
```

```python
n_iter = 0

# StraitifiedKFold의 split( ) 호출시 반드시 레이블 데이터 세트도 추가 입력 필요
for train_index, test_size in skfold.split(features, label):
    
    # split( )으로 반환된 인덱스를 이용해 학습용, 겁증용 테스트 데이터 추출
    x_train, x_test = features[train_index], features[test_size]
    y_train, y_test = label[train_index], label[test_size]
    
    # 학습 및 예측
    dt_clf.fit(x_train, y_train)
    pred = dt_clf.predict(x_test)
    
    # 정확도
    n_iter += 1
    accuracy = np.round(accuracy_score(y_test, pred))
    train_size = x_train.shape[0]
    test_size = x_test.shape[0]
    print('교차 검증 {0}\n 학습데이터 크기: {1}\n 검증데이터 크기: {2}\n 정확도: {3}\n'.format(n_iter, train_size, test_size, accuracy))
    
    cv_accuracy.append(accuracy)
```
